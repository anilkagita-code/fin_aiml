```python
from fyers_apiv3 import fyersModel
```

```python


# client_id = "XC4XXXXM-100"
# access_token = "eyJ0eXXXXXXXX2c5-Y3RgS8wR14g"


client_id = open("fyers_appid.txt", "r").read().strip()
access_token = open("fyers_token.txt", "r").read().strip()
print(client_id)
print(access_token)
# Initialize the FyersModel instance with your client_id, access_token, and enable async mode
fyers = fyersModel.FyersModel(client_id=client_id, is_async=False, token=access_token, log_path="")

data = {
    "symbol":"NSE:SBIN-EQ",
    "resolution":"D",
    "date_format":"0",
    "range_from":"1688389716",
    "range_to":"1691068173",
    "cont_flag":"1"
}

response = fyers.history(data=data)
print(response)
```

```python
# read the response and convert to dataframe
import pandas as pd

candles = response.get('candles', [])
if candles:
    df_candles = pd.DataFrame(candles, columns=['timestamp', 'open', 'high', 'low', 'close', 'volume'])
    df_candles['timestamp'] = pd.to_datetime(df_candles['timestamp'], unit='s')
    df_candles = df_candles.set_index('timestamp')
    print(df_candles.head())
else:
    print("No candle data found in the response.")


```
