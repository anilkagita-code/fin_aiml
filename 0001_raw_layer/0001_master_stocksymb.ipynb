import pandas as pd
import requests
import os
from datetime import datetime

# Step 1: Download the list of NSE stocks (NSE Cash Market CSV)
def download_nse_stock_list():
    url = "https://public.fyers.in/sym_details/NSE_CM.csv"
    df = pd.read_csv(url)
    df.to_csv("NSE_CM_latest.csv", index=False)
    print(f"Downloaded NSE stock list with {len(df)} records")
    return df

# Step 2: Load index constituents (example: NIFTY50)
def load_nifty_index(index_name):
    # Sample URLs or local files for NIFTY indices (replace with actual sources)
    index_urls = {
        "NIFTY50": "https://www.nseindia.com/api/equity-stockIndices?index=NIFTY%2050",
        # Add other indices URLs accordingly
    }
    if index_name in index_urls:
        response = requests.get(index_urls[index_name])
        data = response.json()['data']
        symbols = [item['symbol'] for item in data]
        return set(symbols)
    else:
        return set()

# Step 3: Load FO active stocks list (NSE Futures & Options)
def load_fo_active_stocks():
    # Use official NSE website or API for FO active stocks, here is an example URL
    url = "https://www.nseindia.com/api/fo/fii"
    headers = {
        'User-Agent': 'Mozilla/5.0'
    }
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        data = response.json()
        # Parse symbols from FO data as needed
        fo_symbols = [item['symbol'] for item in data['data']]
        return set(fo_symbols)
    return set()

# Step 4: Get market cap and description for stocks from Fyers API or NSE
def get_stock_details(symbol):
    # Placeholder for actual API call to get market cap and description.
    # Example: call Fyers API or Yahoo Finance here
    # For demo, returning dummy data
    market_cap = 1e10  # Dummy value, in INR
    description = f"Description of {symbol}"
    return market_cap, description

# Main automation function
def automate_nse_stock_processing():
    # Load previously saved stocks to compare for new
    if os.path.exists("NSE_CM_latest.csv"):
        old_df = pd.read_csv("NSE_CM_latest.csv")
        old_symbols = set(old_df['symbol'])
    else:
        old_symbols = set()

    new_df = download_nse_stock_list()
    new_symbols = set(new_df['symbol'])

    # Identify new stocks
    added_symbols = new_symbols - old_symbols
    print(f"New stocks found: {added_symbols}")

    # Load index constituents
    nifty50_symbols = load_nifty_index("NIFTY50")

    # Load FO active stocks
    fo_active_symbols = load_fo_active_stocks()

    # Prepare list for storing stock info
    stocks_info = []

    # Process new stocks (or all if you want full processing)
    for symbol in new_symbols:
        market_cap, description = get_stock_details(symbol)

        # Classify by index
        in_nifty50 = symbol in nifty50_symbols

        # Classify FO active
        fo_active = symbol in fo_active_symbols

        # Filter by market cap threshold (example: > 5000 Cr)
        if market_cap > 5e9:
            stocks_info.append({
                "symbol": symbol,
                "market_cap": market_cap,
                "description": description,
                "in_nifty50": in_nifty50,
                "fo_active": fo_active
            })

    # Save final processed data
    final_df = pd.DataFrame(stocks_info)
    final_df.to_csv("Processed_NSE_Stocks.csv", index=False)
    print(f"Processed {len(final_df)} high market cap stocks.")

if __name__ == "__main__":
    automate_nse_stock_processing()
