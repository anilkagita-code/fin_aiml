<h1>Astronomical Data to Identify pattens using GANNs</h1>

<h3>Approach One</h3>

```python
from PyAstronomy import pyasl
import pandas as pd

dates = pd.to_datetime(['2024-02-01','2024-02-02','2024-02-03','2024-02-04'])
jds = [pyasl.jdcnv(date) for date in dates]
phases = pyasl.moonphase(jds)
print(phases)  # Fraction of illumination [web:22]
```

<h3>Approach Two</h3>

```python
import ephem
import pandas as pd

# Series of dates (assume your stock data dates in 'YYYY-MM-DD' format)
dates = pd.to_datetime(['2024-02-01','2024-02-02','2024-02-03','2024-02-04'])

def get_lunar_phase(date):
    d = ephem.Date(date)
    nnm = ephem.next_new_moon(d)
    pnm = ephem.previous_new_moon(d)
    return (d - pnm)/(nnm - pnm)  # float: 0=new, 0.5=full, 1=new

phases = [get_lunar_phase(date) for date in dates]
print(phases)  # Each value tells you the phase for each date [web:21][web:26]
```
